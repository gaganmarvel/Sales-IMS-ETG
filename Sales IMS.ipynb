```python
import json,datetime

class Products:
    def __init__(self):
        self.items={}
        self.load()
    
    def load(self):
        try:
            jp=open("product.json",'r')
            with jp as jp_in:
                self.items=json.load(jp_in)
                jp.close()
        except:
            jp=open("product.json",'w')
            with jp as jp_in:
                self.items={}
                jp.close()
    
    def __str__(self):
        op=""
        count=0
        l=len(self.items.items())
        print("Total Products: "+str(l))
        print("________________________________________________")
        for id, d in self.items.items():
            count = count + 1
            op=op + str(count)+ "\t"
            op= op + f"PID Number    : {id} \n\tProduct Name  : {d['name']}  \n\tType          : {d['typ']} \n\tPrice         : {d['price']}.00 \n\tQuantity      : {d['qty']}\n"
            op= op + "________________________________________________\n"
        return op
    
    def add(self,Id,name,typ,price,qty):
        self.items[Id]={"name":name,"typ":typ,"price":price,"qty":qty}
        save=self.save()
        if(save):
            return True
    
    def save(self):
        jp=open("product.json",'w')
        with jp as jp_out:
            json.dump(self.items,jp_out)
            jp.close()
            return True
    
    def remove(self,Id):
        del self.items[Id]
        save=self.save()
        if save:
            return True
    
    def search(self,Id):
        prod=self.items.get(Id,None)
        if prod:
            return Id,prod['name'],prod['typ'],prod['price'],prod['qty']
        else:
            return None
    
    def update(self,Id,qt):
        self.items[Id]['qty']-=qt
        self.save()
    
    def updateall(self,Id,k,v):
        if k=='n':
            self.items[Id]['name']=v
        elif k=='t':
            self.items[Id]['typ']=v
        elif k=='p':
            self.items[Id]['price']=v
        elif k=='q':
            self.items[Id]['qty']+=v
        save=self.save()
        if save:
            return True
        
product=Products()

class Sales:
    def __init__(self):
        self.items={}
        self.load()
    
    def load(self):
        try:
            with open("sales.json",'r') as js_in:
                self.items=json.load(js_in)
        except:
            with open("sales.json",'w') as js_in:
                self.items={}
    
    def __str__(self):
        op=""
        count=0
        l=len(self.items.items())
        print("Total Purchase: "+str(l))
        print("\nBill Number\t\t\t|PID\t|Product Name\t\t|Quantity\t\t|Bill Amount")
        print("________________________________________________________________________________________________________________")
        for id, d in self.items.items():
            op= op + f"{id}\t {d['PID']}\t {d['Name']}\t\t\t {d['u_Quantity']}\t\t\t {d['Bill_Amount']}.00\n"
        return op
    
    def add(self,b_no,pid,pname,u_qty,b_amt):
        self.items[b_no]={"PID":pid,"Name":pname,"u_Quantity":u_qty,"Bill_Amount":b_amt}
        save=self.save()
        if(save):
            return True
    
    def save(self):
        with open("sales.json",'w') as js_out:
            json.dump(self.items,js_out)
            return True
    
sale=Sales()

        
```

# Adding Product

```python
print("\n--------------------------------------------------------------------------------------------------")
print("\n***************ADDING PRODUCT***************")
print("____________________________________________")
while True:
    Id=input("Enter PID number for the product: ").upper()
    if product.search(Id):
        print("This PID number is taken, Enter a different PID number")
        continue
    nam=input("Enter the Name of the product: ").upper()
    ty=input("Enter the Type of the product: ").upper()
    pr=int(input("Enter the Price of the product: "))
    quan=int(input("Enter the Quantity of the product: "))
    ad=product.add(Id,nam,ty,pr,quan)
    if(ad):
        print("\n--------Product Added--------\n")
        print("____________________________________________")
        break
```
